# Exercise 00 : Line chart

```python
import pandas as pd
import sqlite3

con = sqlite3.connect(r"../data/checking-logs.sqlite")

query = """
SELECT date(datetime) AS date, COUNT(*) AS views
FROM pageviews
WHERE uid LIKE 'user_%'
GROUP BY date
ORDER BY date
"""

views = pd.read_sql(query, con, parse_dates=['date']).set_index('date')

ax = views.plot(
    figsize=(15, 8),
    legend=False,
    title="Views per day",
    rot=90,                #    90 
    fontsize=8
)

con.close()
# views.head()
print(f"The dataframe for the graph contains, rows: {len(views)}")
```
