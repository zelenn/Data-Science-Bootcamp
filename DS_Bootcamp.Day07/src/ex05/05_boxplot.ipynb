# Exercise 05 : Boxplot

```python
import pandas as pd
import matplotlib.pyplot as plt
# import numpy as np

plt.rcParams.update({"font.size": 8})

df = pd.read_csv('ab-test.csv')

fig, axes = plt.subplots(1, 2, figsize=(15, 8), sharey=True)

boxprops = {'facecolor': 'none', 'edgecolor': 'green', 'linewidth': 3}
medianprops = {'color': 'red', 'linewidth': 2}
whiskerprops = {'color': 'red', 'linewidth': 1}
capprops = {'color': 'red', 'linewidth': 1}
flierprops = {'marker': 'o', 'markerfacecolor': 'black', 'markersize': 4}

for i, (ax, phase) in enumerate(zip(axes, ['after', 'before'])):
    tmp = df[df['time'] == phase]
    
    bp = tmp.boxplot(
        column='diff', 
        by='group', 
        ax=ax,
        boxprops=boxprops,
        medianprops=medianprops,
        whiskerprops=whiskerprops,
        capprops=capprops,
        flierprops=flierprops,
        patch_artist=True
    )
    
    ax.set_title(phase)
    ax.set_xlabel('[group]')
    ax.set_ylabel('diff (hours)')
    ax.grid(True)
    

fig.suptitle('Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
plt.tight_layout()
plt.show()


# Questions
df = df[df['time'].isin(['before'])]
print(f"The dataframe for the graph contains (rows): {len(df)}")

# Calculate IQR for control BEFORE

# control_before = df[(df['group']=='control') & (df['time']=='before')]['diff']
# q1, q3 = np.percentile(control_before, [25,75])
# iqr = q3 - q1
# print("IQR control before (exact):", iqr)
```
